```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('data.csv') 
data.dropna()
```

```python
# Loại bỏ các dữ liệu trùng
data=data.drop_duplicates(subset=data.columns, keep='first')
```

```python
# Chuyển giá tiền về dưới dạng tỉ lệ 1:1000000000
data['Price']=data['Price']/ 1000000000
```

```python
Sum_car=data.Car.value_counts()>5
True_values=Sum_car[Sum_car==True].index
print(len(True_values))
Sum_car=data[data['Car'].isin(True_values)].copy()
Sum_car
```

```python
Sum_car['Price']=Sum_car['Price']/ 1000000000
```

# Phân tích về Price

```python
ad_dims= (16,10)
fig,ax=plt.subplots(figsize=ad_dims)
sb.distplot(Sum_car.Price)
plt.show()

# Biểu đồ bên dưới thể hiện được độ phân bố về giá tiền trong từng khoảng nhất đinh
# Dựa vào biểu đồ ta thấy, trong số dữ liệu thu thập được thì đa số phân bổ cao trong khoảng từ 500 triệu đến dưới 4 tỉ.
```

```python
# Chỉ lấy những loại xe có giá tiền thấp hơn 10 tỉ vì lượng dữ liệu có giá tiền trên 10 tỉ là quá thấp hoặc không có
# nên không thể dùng để train.
Sum_car=Sum_car[Sum_car.Price<=10]
```

```python
Sum_car
```

```python
visualData=Sum_car.copy()
visualData
```

# Phân tích sự ảnh hưởng của dữ liệu cột đối với giá thành.

```python
# Biểu đồ bên dưới thể hiện được giá trị trung bình về từng khoảng giá tiền với xuất xứ của xe tương ứng. 
fig = plt.figure(2,figsize=(8,6))
plt.subplot(211)
sb.boxplot(data=visualData,x='madeIn',y='Price')
plt.title('madeIn vs Price')
plt.tight_layout()

# Thông qua biểu đồ này ta thấy rằng:
# + Trong khoảng từ giá thành từ trên 500 triệu đến dưới 4 tỉ thì không có sự chênh lệch quá lớn giữa xe Nhập khẩu và
# xe Láp ráp trong nước.
# + Với giá tiền trên 5 tỉ thì nhường như loại xe Nhập khẩu chiếm tuyệt đối.
```

```python
# Biểu đồ bên dưới thể hiện được giá trị trung bình của từng loại xe tương ứng với từng khoảng giá tiền

fig = plt.figure(2,figsize=(8,6))
plt.subplot(211)
sb.boxplot(data=Sum_car,x='carType',y='Price')
plt.title('carType vs Price')
plt.tight_layout()

# Thông qua biểu đồ này ta thấy rằng:
# + Dòng xe SUV là dòng xe được ưa chuộng và phân bổ đều ở các khoảng giá tiền khác nhau
# + Dòng xe Truck là dòng xe có giá thành rẻ nhất cũng như ít được người dùng quan tâm.
```

```python
sb.catplot(data=visualData,x='madeIn',y='Price')
plt.title('madeIn vs Price')

# Biểu đồ bên dưới thể hiện rõ hơn về độ phân bố của từng khoảng giá tiền với xuất xứ của xe tương ứng. 
```

```python

sb.catplot(data=Sum_car,x='carType',y='Price', height=5, aspect=3)
plt.title('carType vs Price')

# Bên dưới thể hiện rõ hơn về độ phân bố của từng loại xe tương ứng với từng khoảng giá tiền.
```
